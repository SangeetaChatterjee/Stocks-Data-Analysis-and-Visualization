# TASK #5: PORTFOLIO ALLOCATION - DAILY RETURN/WORTH CALCULATION (FUNCTION)

```
# Lets assume we have $1,000,000 to be invested and we will allocate this fund based on the weights of the stocks
# We will create a function that takes in the stock prices along with the weights and return:
# (1) Daily value of each individual securuty in $ over the specified time period
# (2) Overall daily worth of the entire portfolio 
# (3) Daily return 

def portfolio_allocation(df, weights):
  df_portfolioalloc = df.copy()
  df_portfolioalloc = data_normalize(df_portfolioalloc)

# Calculating total for each stock-for each day based on the weight and assuming total investment of $1000000
  for counter , stock in enumerate(df_portfolioalloc.columns[1:]):
    df_portfolioalloc[stock] = df_portfolioalloc[stock] * weights[counter]
    df_portfolioalloc[stock] = df_portfolioalloc[stock] * 1000000

# Creating a new column to calculate the daily return for each day
  df_portfolioalloc['Total Daily Returns'] = df_portfolioalloc[df_portfolioalloc != 'Date'].sum(axis = 1)

#Creating another column to calculate the total percentage of daily returns for each day
  df_portfolioalloc['Daily % Returns'] = 0.000
  for i in range(1, len(df_portfolioalloc)):
    df_portfolioalloc['Daily % Returns'][i] = ( (df_portfolioalloc['Total Daily Returns'][i] - df_portfolioalloc['Total Daily Returns'][i-1]) / df_portfolioalloc['Total Daily Returns'][i-1]) * 100
  
  df_portfolioalloc['Daily % Returns'][0] = 0 
  return df_portfolioalloc

```

**MINI CHALLENGE #4:**
- **Call the function and ensure that the results make sense**

```
df_portfolio = portfolio_allocation(stock_df,weights)
df_portfolio
```
